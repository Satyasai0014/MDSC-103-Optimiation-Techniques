```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(a):
    return 4*a[0] + 6*a[1] - 2*a[0]**2 - 2*a[0]*a[1] - 2*a[1]**2

def gradf(a):
    return np.array([4-4*a[0]-2*a[1], 6-2*a[0]-4*a[1]])

x1 = np.linspace(0,10,100)
x2 = np.linspace(0,10,100)

X,Y = np.meshgrid(x1,x2)
C = (X,Y)
F = f(C)
fig = plt.figure(figsize = (10,10))
ax = plt.subplot(projection = '3d')
surf = ax.plot_surface(X,Y,F)
ax.contour(X,Y,F)
ax.scatter3D(0.33300781,1.33300781,f((0.33300781,1.33300781)),color = 'blue')

```

```python
a1 = float(input("Enter Initial Point 1 (x1) :"))
a2= float(input("Enter Initial Point 2 (x2) :"))

a = np.array([a1,a2])

step = float(input("Enter Learning Rate (r) :"))
itr = int(input("Enter no. of iterations :"))
n = itr



x1 = np.array([0,0])
err = np.array([0.2,1.200])

while(itr):
    
    if(not np.allclose(gradf(a), err)):
        
        print("\nIteration",abs(itr-n))
        print(f"Grad f at x1={a[0]} and x2={a[1]} is : {gradf(a)}\n")
        x1 = a + step*gradf(a)
        a = x1
        itr = itr-1
        
    else:
        break        
print("\nThe Solution After Iterations is :\n")
print("Critical Point(X*) :",a)
print("\n f(X*) at X* = ",f(a))
```
