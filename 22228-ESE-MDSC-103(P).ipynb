```python
import matplotlib.pyplot as plt
import numpy as np


f=lambda x1,x2: 4*x1+6*x2-2*x1**2-2*x1*x2-2*x2**2

```

```python
import numpy as np


x=np.linspace(-2,1,10)
y=x.copy()
z = []
for i in range(0,len(x)):
    for j in range(i,len(y)):
        z.append(f(x[i],y[j]))
plt.plot(x,y,z)
plt.show()
```
