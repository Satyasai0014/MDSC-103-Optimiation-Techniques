# $$ Golden Section$$


```python
# importing Required Libraries 
import numpy as np
```

```python
#Author: A.Satya Sai
#Date: 14-09-2022

# This program is about Golden Section in NLPP 

def f(x): # Here we are defining function with some conditions
    if (0<= x <= 2):
        a= lambda x : 4*x
        #plt.plot(x,a(x))
        return a(x)
    if (2<=x<=4):
        b = lambda x: 4-x
        #y=[]
        #q= np.linspace(0,4,100)
        #y.append(x)
        #plt.plot(x,b(x))
        return b(x)
    
delta = (np.sqrt(5)-1)/2  # the is called the Golden ratio value which  will be used in x1 and x2 calculation      
def golden(c,d,p): # this the main function .
    xl=c # This is the starting limit of function
    xr =d # This is the Ending Limit of the function
    x1=0
    x2 = 0
    #a = 0
    p = int(input("Enter the number of Iterations: "))
    for i in range(p):
        x1=xr-delta*(xr-xl) #Formula to calculate x1 value in the Golden Section method of NLPP
        x2=xl+delta*(xr-xl)  #Formula to calculate x2 value in the Golden Section method of NLPP
        
        
        # From here the conditions will start ..where it will check and change the xl and xr value after each Iteration.
        if f(x1)>f(x2):
            xr=x2
        elif f(x1)<f(x2):
            xl=x1
        elif f(x1)==f(x2):
            xl=x1
            xr=x2
        
        print("")
        print(f'The values for xl,xr after {i} Iteration is [{x1},{x2}]')
        print("___________________________________________________________________")
        print("")
    print("***********************************************************************")
    print(f"THE VALUES OF xl,xr AFTER {i} ITERATION IS [{x1},{x2}]")
    print("***********************************************************************")

c = int(input("Enter Starting limit of function: "))
d = int(input("Enter Ending  limit of function: "))
golden(0,4,10)

   
```
