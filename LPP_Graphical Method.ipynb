```python
## IMPORTING REQUIRED LIBRARIES

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
from math import *
from sympy import symbols, Eq, solve

```

```python
p = np.linspace(-2,15,1000)
q = np.linspace(-2,15,1000)

x,y = np.meshgrid(q,q)

def con1(x):
        return (24-6*x)/4   
def con2(x):
        return (6-x)/2    
def con3(x):
        return 1+x
def con4(x):
        return 2+x-x
def con5(x):
        return 0+x-x
    

figure(figsize=(20,9), dpi=80)
plt.plot(p,con1(p),c='r',label = "6x+4y<=24")
plt.plot(p,con2(p),c='g',label = "x+2y<=6")
plt.plot(p,con3(p),c='b',label = "-x+y<=1")
plt.plot(p,con4(p),c='orange',label = "y<=2")
plt.plot(p,con5(p),c='black',label = "x>=0")
plt.plot(0+p-p,p*2,c='black',label = "y>=0")

plt.title("Reddy Mikks Problem Using Graphical Method",fontsize = 16)
plt.xlabel("X -Axis-------------->",fontsize = 15)
plt.ylabel("Y -Axis-------------->",fontsize = 15)

plt.imshow(((6*x+4*y<=24) & (x+2*y<=6) & (-x+y<=1) & (y<=2)) & (x>=0) & (y>=0).astype(int) , 
                extent=(x.min(),x.max(),y.min(),y.max()),origin="lower", cmap="GnBu", alpha = 0.39);
plt.legend(loc=5)
plt.show()
```

```python
x, y = symbols('x,y')
  

eq1 = Eq((6*x+4*y), 24)

eq2 = Eq((x+2*y), 6)

eq3 = Eq((-x+y), 1)

eq4 = Eq((y), 2)

eq5 = Eq((x) , 0)

eq6 = Eq((y), 0)

  

print("Values of 2 unknown variable are as follows:")
  
sol1=solve((eq1, eq2), (x, y))


sol2=solve((eq1, eq3), (x, y)) 
sol3=solve((eq1, eq4), (x, y)) 
sol4=solve((eq1, eq5), (x, y)) 
sol5=solve((eq1, eq6), (x, y)) 

sol6=solve((eq2, eq3), (x, y)) 
sol7=solve((eq2, eq4), (x, y)) 
sol8=solve((eq2, eq5), (x, y)) 
sol9=solve((eq2, eq6), (x, y))

sol10=solve((eq3, eq4), (x, y))
sol11=solve((eq3, eq5), (x, y)) 
sol12=solve((eq3, eq6), (x, y)) 


sol13=solve((eq4, eq5), (x, y)) 

sol14=solve((eq5, eq6), (x, y)) 
```
